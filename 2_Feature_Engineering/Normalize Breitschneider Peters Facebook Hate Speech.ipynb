# Normalize

the goal is to develop a simple normalised form of hate speech datasets when combining different sources. For now, I include the following values:

- raw text: the comment unprocessed
- severity: the average of all expert assessments with a value between 0 and 1
- binary_label: A binary estimation of the comment on beeing hate speech or not

```python
# calculate severity
def calculateSeverity(val):
    value = val.split(',') if isinstance(val, str) else False
    if(value):
        # average of all experts divided by highest possible value (2)
        return int(sum(map(int,value)) / len(value)) / 2
    else:
        return 0
df['severity'] = df['valence'].apply(lambda x: calculateSeverity(x))
df.severity.unique()
```

```python
# generate binary label from severity
df['binary_label'] = df['valence'].apply(lambda x: calculateSeverity(x) >= 0.5)
```

```python
print("Out of {} rows, {} are ordinary, {} are hate speech of some kind".format(len(df),
                                                      len(df[df['binary_label'] == False]),
     len(df[df['binary_label'] == True])))
```

```python
df_hs = df.filter(['comment', 'severity', 'binary_label'], axis=1)
df_hs.rename(columns={'comment':'raw_text'}, inplace=True)
df_hs.head(20)
```

```python
df_hs.to_csv('data/normalized_facebook_comments.csv', index=False)
df_hs.to_excel('data/normalized_facebook_comments.xlsx', index=False)
```
