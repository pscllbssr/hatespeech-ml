# Data preparation

```python
#df.valence = df.valence.astype(str)
df['label'] = df['valence'].apply(lambda x: sum(x.split(',')))
```

```python
def computeLabel(val):
    value = val.split(',') if isinstance(val, str) else False
    if(value):
        return int(sum(map(int,value)) / len(value))
    else:
        return 0
```

```python
for index, row in df[:10].iterrows():
    print(computeLabel(row.valence))    
```

```python
df['label'] = df['valence'].apply(lambda x: computeLabel(x))
```

```python
df.head()
```

```python
df['binary_label'] = df['valence'].apply(lambda x: computeLabel(x) > 0)
df.head()
```

```python
df_hs = df.filter(['comment', 'label', 'binary_label'], axis=1)
df_hs.head()
```

```python
df_hs[:200]
```

```python
import string 
import nltk
import re
stopwords = nltk.corpus.stopwords.words('german') 
lm = nltk.WordNetLemmatizer()
```

```python
def clean_input(text):
    # remove user hashes in comment text with "@" and without
    
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [lm.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

```python
for index, row in df[:10].iterrows():
    print(clean_input(row.comment))
```

```python
# --> eventuell hier ein problem, da gleiche wörter aber gegensätzliche meinung:
print(df_hs.iloc[0]['comment'] + '\n' + df_hs.iloc[3]['comment'])
```

# Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vect = TfidfVectorizer(analyzer=clean_input)
X_tfidf = tfidf_vect.fit_transform(df_hs['comment'])
print(X_tfidf.shape)
print(tfidf_vect.get_feature_names())
```

```python
X_tfidf_df = pd.DataFrame(X_tfidf.toarray())
X_tfidf_df.columns = tfidf_vect.get_feature_names()
X_tfidf_df[:50]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_tfidf_df, df_hs['binary_label'], test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, max_depth=20, n_jobs=-1)
rf_model = rf.fit(X_train, y_train)
```

```python
sorted(zip(rf_model.feature_importances_, X_train.columns), reverse=True)[:10]
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
```

```python
y_pred = rf_model.predict(X_test)
precision, recall, fscore, support = score(y_test, y_pred, pos_label=True, average='binary') # pos_label= wenn nicht binary, dann label angeben
```

```python
print('Precision: {} / Recall: {} / Accuracy: {}'.format(round(precision, 3), 
                                                         round(recall, 3), 
                                                         round((y_pred==y_test).sum() / len(y_pred))))
```

```python
def estimate_input(text_array):
    text_vect = tfidf_vect.transform(text_array)
    cls = rf_model.predict(text_vect.toarray())
    prob = rf_model.predict_proba(text_vect.toarray())
    return (cls, prob)
          
```
