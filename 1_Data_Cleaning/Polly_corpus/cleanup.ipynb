```python
import pandas as pd
```

```python
df = pd.read_csv('data/POLLY_tweets_with_hate.csv')
```

```python
df.info(verbose=True)
```

```python
df.head()
```

```python
df['binary_label'] = True
```

```python
df
```

```python
# remove entries with image reference as context is more difficult to understand
df = df[df['Image'].isna()].filter(items=['Tweet', 'binary_label'])
df
```

```python
# remove usernames
import re
def cleanup_tweet(tweet):
    tweet = re.sub(r'http\S+', '', tweet)
    tweet = re.sub(r"@(\w+)", '', tweet)
    tweet = re.sub(r"RT", '', tweet)
    return tweet
print(cleanup_tweet(df.iloc[1]['Tweet']))
print(cleanup_tweet("RT @name: Wenn R eine rt Schule schon wegen der rel"))
```

```python
df['raw_text'] = df['Tweet'].apply(lambda x: cleanup_tweet(x))
df
```

```python
df['severity'] = 1
```

```python
df = df.filter(items=['raw_text', 'severity', 'binary_label'])
df
```

```python
df.to_csv('../../2_Feature_Engineering/data/polly_hatespeech.csv', index=False)
df.to_excel('../../2_Feature_Engineering/data/polly_hatespeech.xlsx', index=False)
```
