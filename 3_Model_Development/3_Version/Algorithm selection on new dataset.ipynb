```python
import pandas as pd
```

```python
df = pd.read_csv('../../2_Feature_Engineering/export/randomized_balanced_dataset.csv')
df
```

# Clean text

```python
from model_helpers import clean_input as ci
```

```python
df['cleaned_text'] = df['raw_text'].apply(ci)
```

```python
df
```

# Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_bigram = TfidfVectorizer(ngram_range=(2,2))
tfidf_simple = TfidfVectorizer()
```

```python
X_features_bigram = tfidf_bigram.fit_transform(df['cleaned_text'])
X_features_bigram
```

```python
X_features_simple = tfidf_simple.fit_transform(df['cleaned_text'])
X_features_simple
```

```python
tfidf_trigrams = TfidfVectorizer(ngram_range=(3,3))
X_features_trigrams = tfidf_trigrams.fit_transform(df['cleaned_text'])
X_features_trigrams
```

# Train with K-Fold & Random Forest

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
k_fold = KFold(n_splits=5, shuffle=True)
```

```python
# k-fold without n-grams
cross_val_score(rf, X_features_simple, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
# k-fold with bigrams
cross_val_score(rf, X_features_bigram, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
# k-fold with trigrams
cross_val_score(rf, X_features_trigrams, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

# Train with K-Fold & Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
# logistic regression without n-grams
cross_val_score(lr, X_features_simple, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
# k-fold with bigrams
cross_val_score(lr, X_features_bigram, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

# Train with K-Fold & Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
# NB without n-grams
cross_val_score(nb, X_features_simple.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
# NB with bigrams
cross_val_score(nb, X_features_bigram.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

# Train with K-Fold & Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
# MNB without n-grams
cross_val_score(mnb, X_features_simple.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
# MNB with bigrams
cross_val_score(nb, X_features_bigram.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

# Train with K-Fold & support vector machine

```python
from sklearn.linear_model import SGDClassifier
```

```python
svm = SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, random_state=42, max_iter=5, tol=None) # params only copied
```

```python
# SVM without n-grams
cross_val_score(svm, X_features_simple.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
# SVM with bigrams
cross_val_score(svm, X_features_bigram.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

# results

seems as using n-grams would not improve the model. following the best from all algorithms

```python
svm = cross_val_score(svm, X_features_simple.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
mnb = cross_val_score(mnb, X_features_simple.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
gnb = cross_val_score(nb, X_features_simple.toarray(), df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
lr = cross_val_score(lr, X_features_simple, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
rf = cross_val_score(rf, X_features_simple, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)

print('scores')
print('svm: {}'.format(svm))
print('mnb: {}'.format(mnb))
print('gnb: {}'.format(gnb))
print('lr: {}'.format(lr))
print('rf: {}'.format(rf))
```

```python
print('score averages')
print('svm: {}'.format(sum(svm) / len(svm)))
print('mnb: {}'.format(sum(mnb) / len(mnb)))
print('gnb: {}'.format(sum(gnb) / len(gnb)))
print('lr: {}'.format(sum(lr) / len(lr)))
print('rf: {}'.format(sum(rf) / len(rf)))
```
