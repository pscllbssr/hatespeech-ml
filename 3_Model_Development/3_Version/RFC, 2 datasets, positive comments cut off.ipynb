```python
import pandas as pd
```

```python
df = pd.read_csv('../../2_Feature_Engineering/export/randomized_balanced_dataset.csv')
df
```

# Clean text

```python
from model_helpers import clean_input as ci
```

```python
df['cleaned_text'] = df['raw_text'].apply(ci)
```

```python
df
```

# Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(ngram_range=(2,2))
```

```python
X_features = tfidf.fit_transform(df['cleaned_text'])
X_features
```

# Train with K-Fold

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
k_fold = KFold(n_splits=5, shuffle=True)
cross_val_score(rf, X_features, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=1)
```

```python
k_fold = KFold(n_splits=5, shuffle=True)
cross_val_score(rf, X_features, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=-1)
```

```python
k_fold = KFold(n_splits=5, shuffle=True)
cross_val_score(rf, X_features, df['binary_label'], cv=k_fold, scoring='accuracy', n_jobs=10)
```
