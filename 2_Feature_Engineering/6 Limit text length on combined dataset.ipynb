```python
import pandas as pd
```

```python
df = pd.read_csv(r"export/randomized_balanced_dataset.csv")
```

```python
df.head()
```

```python
total = len(df)
twitter_length = len(df[df['raw_text'].str.len() <= 280])
medium = len(df[df['raw_text'].str.len() > 280])
print(total, twitter_length, medium)
```

```python
hate_df = df[df['binary_label'] == True]
love_df = df[df['binary_label'] == False]
```

```python
def divide_on_length(df, threshold):
    total = len(df)
    shorter = len(df[df['raw_text'].str.len() <= threshold])
    longer = len(df[df['raw_text'].str.len() > threshold])
    print(total, shorter, longer)
```

```python
print('hate dataframe: total, fits in, too long')
divide_on_length(hate_df, 280)
print()

print('love dataframe: total, fits in, too long')
divide_on_length(love_df, 280)
print()
```

```python
print('hate dataframe: total, fits in, too long')
divide_on_length(hate_df, 300)
print()

print('love dataframe: total, fits in, too long')
divide_on_length(love_df, 300)
print()
```

```python
print('hate dataframe: total, fits in, too long')
divide_on_length(hate_df, 500)
print()

print('love dataframe: total, fits in, too long')
divide_on_length(love_df, 500)
print()
```

```python
# divide
threshold = 300
hate_df = hate_df[hate_df['raw_text'].str.len() <= threshold]
love_df = love_df[love_df['raw_text'].str.len() <= threshold]
```
