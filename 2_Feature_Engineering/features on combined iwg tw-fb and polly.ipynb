```python
from model_helpers import clean_input as ci
```

```python
df['cleaned_text'] = df['raw_text'].apply(ci)
df
```

```python
# remove usernames + RT
import re
def cleanup_tweet(tweet):
    tweet = re.sub(r"@(\w+)", '', tweet)
    tweet = re.sub(r"RT", '', tweet)
    return tweet
df['cleaned_text'] = df['cleaned_text'].apply(cleanup_tweet)
df
```

```python
def extract_keywords(text):
    keywords = re.findall(r"#(\w+)", text)
    return ",".join(keywords)
extract_keywords(df.iloc[10]['raw_text'])
```

```python
df['keywords'] = df['raw_text'].apply(extract_keywords)
df
```
