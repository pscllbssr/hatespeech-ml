# punctuation

```python
import string
string.punctuation
```

```python
type(string.punctuation)
```

```python
def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation.replace('#', '')]) # without gartenzaun as it indicates a hashtag
    return round(count/(len(text) - text.count(" ")),3) * 100
```

```python
df['punct%'] = df['raw_text'].apply(lambda x: count_punct(x))
df.head()
```

```python
df.head(-10)
```

```python
bins = np.linspace(0, 50, 40)

pyplot.hist(df[df['binary_label'] == True]['punct%'], bins, alpha=0.5, normed=True, label='hate speech')
pyplot.hist(df[df['binary_label'] == False]['punct%'], bins, alpha=0.5, normed=True, label='no hate speech')

pyplot.legend(loc='upper right')
pyplot.show()
```

```python
from textblob_de import TextBlobDE as TextBlob
```

```python
def sentiment_analysis_pol(text):
    blobber = TextBlob(text)
    return blobber.sentiment.polarity
sentiment_analysis_pol(df.iloc[0]['raw_text'])
```

```python
def sentiment_analysis_sub(text):
    blobber = TextBlob(text)
    return blobber.sentiment.subjectivity
sentiment_analysis_sub(df.iloc[0]['raw_text'])
```

```python
for index in range(0,5):
    print(sentiment_analysis_pol(df.iloc[index]['raw_text']))
    print(sentiment_analysis_sub(df.iloc[index]['raw_text']))
    print()
```

```python
df['sent_polarity'] = df['raw_text'].apply(lambda x: sentiment_analysis_pol(x))
df['sent_subjectivity'] = df['raw_text'].apply(lambda x: sentiment_analysis_sub(x))
```

```python
df.head()
```

```python
bins = np.linspace(-1, 1, 20)

pyplot.hist(df[df['binary_label'] == True]['sent_polarity'], bins, alpha=0.5, normed=True, label='hate speech')
pyplot.hist(df[df['binary_label'] == False]['sent_polarity'], bins, alpha=0.5, normed=True, label='no hate speech')

pyplot.legend(loc='upper right')
pyplot.show()
```

```python
bins = np.linspace(0, 1, 20)

pyplot.hist(df[df['binary_label'] == True]['sent_subjectivity'], bins, alpha=0.5, normed=True, label='hate speech')
pyplot.hist(df[df['binary_label'] == False]['sent_subjectivity'], bins, alpha=0.5, normed=True, label='no hate speech')

pyplot.legend(loc='upper right')
pyplot.show()
```
