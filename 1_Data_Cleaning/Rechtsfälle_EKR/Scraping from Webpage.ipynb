```python
import requests
from bs4 import BeautifulSoup as BS
```

```python
r  = requests.get(urls[0], params={"db":"N", "keyword4": "495", "p":"1"})
print(r)
```

```python
soup = BS(r.text)
```

```python
print(soup)
```

```python
# find all links in table, as there is only one table
table = soup.find('table')
print(table)
```

```python
links = set([])
for link in table.find_all('a'):
    links.add(link['href'])
print(links)
```

```python
print(len(links))
```

```python
# now grab all urls
for url in urls:
    r  = requests.get(urls[0])
    soup = BS(r.text)
    table = soup.find('table')
    for link in table.find_all('a'):
        links.add(link['href'])
print(len(links))
```

```python
from urllib.parse import urlparse

def evaluate_link(link):
    link_obj = urlparse(link)
    path_list = list(filter(None, link_obj.path.split('/')))
    if len(path_list) == 3:
        print('true: {}'.format(path_list))
        return True
    else:
        print('false: {}'.format(path_list))
        return False
```

```python
for link in links:
    evaluate_link(link)
```

```python
cleaned_links = [link for link in links if evaluate_link(link)]
print(cleaned_links)
print(len(cleaned_links))
```

```python
base_url = r"http://www.ekr.admin.ch"
case_html_list = []
```

```python
successful_links = []
for link in cleaned_links:
    
    resp = requests.get(base_url + link)
    print(resp.encoding)
    # content is improperly detected as Latin-1, but is utf-8
    
    try:
        soup = BS(resp.content.decode('utf-8'))
        case_html_list.append(soup)
        successful_links.append(link)
    except:
        # skip
        print("skipped {}".format(link))
```

```python
# improve above function in case there are more p-tags than headlines (h3)
def extract_content(html, link):
    
    case = {}
    case['link'] = base_url + link
    
    # get only relevant content (index 8 is hard coded)
    content = html.find_all('div', attrs={'class': 'container-fluid'})    
    main_content = content[8]
    
    # extraxt text blocks with titles
    text_elements = main_content.find_all(['p', 'h3'], attrs={'class': None})
    text = ""
    
    # all text as one string, headlines marked with "#"   
    for tag in text_elements:
        if(tag.name == 'h3'):
            text += "#" + tag.text + "\n"
        else:
            text += tag.text + "\n"
        
    case['text'] = text
        
    # extract additional info from tables (all hard coded indizes)
    tables = main_content.find_all('table')
    case['year'] = tables[0].find_all('tr', attrs={'role': None})[0].find_all('td')[0].text
    case['result'] = tables[0].find_all('tr', attrs={'role': None})[0].find_all('td')[2].text
    
    for row in tables[2].find_all('tr'):
        cells = row.find_all('td')
        if(len(cells) == 2):
            case[cells[0].text] = cells[1].text
    
    #print(case)
    return case
print(extract_content(case_html_list[11], successful_links[11]))
```
